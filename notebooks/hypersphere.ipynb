```python
import torch
import numpy as np
from functools import partial
from math import pi, gamma
import matplotlib.pyplot as plt
```

Surface area

```python
def unit_hypersphere_surface(m):
    return 2 * pi ** (m / 2) / gamma(m / 2)

dims = range(1, 100)
surface_areas = list(map(unit_hypersphere_surface, dims))
plt.plot(dims, surface_areas)
plt.xticks([0, 5, 10, 15, 20, 40, 60, 80, 100])
plt.xlabel('Dimension')
plt.ylabel('Surface Area')
plt.title('Surface area of a unit hypersphere (r=1)')
```

```python
def hypersphere_surface(r, d):
    return (r ** d) * 2 * pi ** (d / 2) / gamma(d / 2)

radius = range(1, 200)
dims = [2, 4, 8, 32, 128]
surface_areas = np.array(
    [list(map(partial(hypersphere_surface, d=d), radius)) for d in dims]
)
plt.plot(radius, np.log(surface_areas.T), label=dims)
plt.legend()
# plt.xticks([0, 5, 10, 15, 20, 40, 60, 80, 100])
plt.xlabel('Radius')
plt.ylabel('Log surface area')
# plt.title('Surface area of a unit hypersphere (r=1)')
```
